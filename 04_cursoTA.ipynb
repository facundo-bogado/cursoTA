```python
datos = json.loads(r.text)
datos
```

```python
fecha = datos['chart']['result'][0]['timestamp']
```

```python
fecha
```

```python
import datetime
```

```python
fecha_parseado = []
for e in fecha:
    d = datetime.datetime.utcfromtimestamp(e).strftime('%d-%m-%Y %H:%M:%S')
    fecha_parseado.append(d)
print(fecha_parseado)
```

```python
fecha = fecha_parseado
```

```python
precios_volumen = datos['chart']['result'][0]['indicators']['quote']
print(precios_volumen)
```

```python
minimo = precios_volumen[0]['low']
minimo
```

```python
volumen = precios_volumen[0]['volume']
maximo = precios_volumen[0]['high']
apertura = precios_volumen[0]['open']
cierre = precios_volumen[0]['close']

print(cierre)
```

```python
preplanilla = dict()
preplanilla['apertura'],preplanilla['maximo'],preplanilla['minimo'],preplanilla['cierre'],preplanilla['volumen']=apertura,maximo,minimo,cierre,volumen
preplanilla
```

```python
planilla = pandas.DataFrame(preplanilla,columns=['apertura','maximo','minimo','cierre','volumen'],index=fecha)
```

```python
planilla = planilla.dropna()
planilla
```
